```
library(SportsAnalytics)
```

```
NBA0809<-fetch_NBAPlayerStatistics("08-09")
```

```
str(NBA0809)
```

```
#利用長條圖比較NBA0809的隊伍中達成三分球的比例
ggplot(data=NBA0809,aes(x=Team,y=ThreesMade))+geom_col()
```

```
ggplot(data=NBA0809,aes(x=Team,y=ThreesMade,fill=Team))+geom_col()
```

```
#利用散佈圖比較NBA0809的隊伍中搶球的比例
ggplot(data=NBA0809,aes(x=Steals,y=Team))+geom_point()
```

```
ggplot(data=NBA0809,aes(x=Steals,y=Team))+geom_point(colour="orange")
```

```
#利用線圖比較NBA0809的隊伍中被取消資格的比例
ggplot(data=NBA0809,aes(x=Team,y=Disqualifications))+geom_line()
```

```
ggplot(data=NBA0809,aes(x=Team,y=Disqualifications))+geom_line(colour="brown")
```

```
#利用盒狀圖比較每個位置達成三分球的比例
ggplot(data=NBA0809,aes(x=Position,y=ThreesMade))+geom_boxplot()
```
